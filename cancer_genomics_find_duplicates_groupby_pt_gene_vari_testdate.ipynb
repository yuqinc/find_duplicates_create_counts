```python
result_list=[]

for ind in df1_pan.index:   
#for ind in [0,1]:   
    
    df1_row=df1_pan[df1_pan.index==ind]
    
    A=df1_row.person_id.to_list()[0]
    B=df1_row.gene.to_list()[0]
    C=df1_row.variant.to_list()[0]
    D=df1_row.tested_date.to_list()[0]
        
    dfgrouped_subset=df1_pangroup[df1_pangroup.person_id==A][df1_pangroup.gene==B][df1_pangroup.variant==C][df1_pangroup.tested_date==D]
    #count=dfgrouped_subset.counts.to_list()[0]
    count=dfgrouped_subset.counts.to_list()[0]
    
    print(ind,A,B,C,D,count)
    result_per_pt=[ind,A,B,C,D,count]
    result_list.append(result_per_pt)
```

```python
df_results=pd.DataFrame(result_list)
df_results.columns=(['index',"person_id","gene","variant","tested_date",'duplicate_count'])
print(shape(df_results))
```
